# Testes com - Adaptive Thresholding

```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt

imagem = cv2.imread('R2_IN.jpg')  

# Imagem Original
image = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
# Imagem em Escala de Cinza
img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# threshold Binario
ret, th1 = cv2.threshold(img,160, 255, cv2.THRESH_BINARY) 

# Adaptativo Threshold - Media
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,
            cv2.THRESH_BINARY,11,2)
# Adaptativo Threshold - Gaussiano
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
            cv2.THRESH_BINARY,11,2)
```

```python
plt.imshow(image)
plt.title("ORIGINAL") 
plt.show()
```

```python
plt.imshow(th1)
plt.title("THRESH - BINARIO")   
plt.show()
```

```python
plt.imshow(th2)
plt.title("ADAPTIVE THRESH - MEDIA") 
plt.show()
```

```python
plt.imshow(th3)
plt.title("ADAPTIVE THRESH - GAUSSIANO") 
plt.show()
```

# Testes com - Matching

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('R2_IN.jpg',0)
img2 = img.copy()
template = cv.imread('R2_EMB.jpg',0)
w, h = template.shape[::-1]

# Aplicando Matching
res = cv.matchTemplate(img,template,eval('cv.TM_CCOEFF'))
min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)

top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv.rectangle(img,top_left, bottom_right, 255, 2)

plt.subplot(121)
plt.imshow(res,cmap = 'brg')
plt.title('Matching Result')

plt.subplot(122)
plt.imshow(img,cmap = 'brg')
plt.title('Detected Point')

plt.show()
```
