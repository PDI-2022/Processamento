```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

# TESTE 1
### Amostra de Branco Leitoso Externa

```python
path='/content/Test1.png'
```

```python
img = cv2.imread(path)

plt.axis("off")

leitoso1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(leitoso1)
plt.show()
```

```python
# histograma da imagem RGB e para cada componente 
hist_red = cv2.calcHist([leitoso1],[0],None,[256],[0,256])
hist_green = cv2.calcHist([leitoso1],[1],None,[256],[0,256])
hist_blue = cv2.calcHist([leitoso1],[2],None,[256],[0,256])

plt.figure(figsize=(16, 4))
plt.xlim([0,256])
plt.subplot(141); plt.plot(hist_red,color='red',lw=2)
plt.subplot(142); plt.plot(hist_green,color='green',lw=2)
plt.subplot(143); plt.plot(hist_blue,color='blue',lw=2)

plt.show()
```

```python
color = ('r','g','b')

for i,col in enumerate(color):
    histr = cv2.calcHist([leitoso1],[i],None,[256],[0,256])
    plt.plot(histr,color=col,lw=2)
    plt.xlim([0,256])
plt.grid()
plt.show()
```

```python
print("Media:", np.average(leitoso1))
print("Media - Hist:", np.average(histr), "\n")

print("Desvio Padro" , np.std(leitoso1))
print("Desvio Padro - Hist" , np.std(histr))
```

```python
path_gray = 'image_gray.png'
image_gray = cv2.imread(path_gray)
plt.imshow(image_gray)
plt.show()
```

```python
hist = cv2.calcHist([image_gray],[0],None,[256],[0,256])
plt.plot(hist,color='black',lw=2)
plt.show()
```

```python
print("Media:", np.average(image_gray))
print("Media - Hist:", np.average(hist),"\n")

print("Desvio Padro" , np.std(image_gray))
print("Desvio Padro - Hist" , np.std(hist))
```

# Teste 4
### Equalizao

```python
imagem_gray = cv2.imread('R2_INTERNA_gray.png')
imgGray = cv2.cvtColor(imagem_gray, cv2.COLOR_BGR2GRAY)
```

```python
hist = cv2.calcHist([imgGray],[0],None,[256],[0,256])
imgHistEq = cv2.equalizeHist(imgGray)

histEq = cv2.calcHist([imgHistEq],[0],None,[256],[0,256])

plt.figure(figsize=(12, 4))
plt.subplot(121); 
plt.plot(hist)
plt.title("Histograma")

plt.subplot(122); 
plt.plot(histEq)
plt.title("Histograma Equalizado")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(121); 
plt.axis("off"); 
plt.imshow(imgGray, cmap = 'gray')

plt.subplot(122); 
plt.axis("off"); 
plt.imshow(imgHistEq, cmap = 'gray') 

plt.show()
```

# Teste 5
### Equalizao - CLAHE

```python
imgCLAHE = cv2.imread('R2_INTERNA_gray.png',0)
```

```python
clahe = cv2.createCLAHE(clipLimit = 20, tileGridSize = (4, 4))
equalized = clahe.apply(imgCLAHE)

# Equalizao do histograma
imgHistEq = cv2.equalizeHist(imgCLAHE)

fig = plt.figure(figsize=(12, 6))

plt.subplot(131); 
plt.imshow(imgCLAHE,cmap='gray') # Original
plt.title("Original")

plt.subplot(132); 
plt.imshow(equalized,cmap='gray') # CLAHE
plt.title("CLAHE")

plt.subplot(133); 
plt.imshow(imgHistEq,cmap='gray') # Equalizao do histograma
plt.title("Equalizao do histograma")
plt.show()
```

```python
hist_original = cv2.calcHist([imgCLAHE],[0], None, [256], [0, 256])
histClahe = cv2.calcHist([equalized],[0],None,[256],[0,256])
histEq = cv2.calcHist([imgHistEq],[0],None,[256],[0,256])



fig = plt.figure(figsize=(12, 6))
plt.subplot(131); 
plt.plot(hist_original)
plt.title("Original")
plt.subplot(132); 
plt.plot(histClahe)
plt.title("CLAHE")
plt.subplot(133); 
plt.plot(histEq)
plt.title("Equalizao do histograma")
plt.show()
```
