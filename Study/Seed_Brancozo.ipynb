```python
brancozo = cv2.imread('LOTE_10_R1_EXTERNA.jpg')

plt.axis("off")
median = cv2.medianBlur(brancozo,5)
median = cv2.cvtColor(brancozo, cv2.COLOR_BGR2RGB)
plt.imshow(median)
plt.show()
```

```python
median = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(median, cv2.COLOR_BGR2HSV)

lower = np.array([0, 0, 168])
upper = np.array([172, 111, 255])

mask = cv2.inRange(hsv, lower, upper)
plt.axis("off")
plt.imshow(mask)
plt.show()
```

```python
median = cv2.cvtColor(median, cv2.COLOR_BGR2RGB)

result = cv2.bitwise_and(median ,median, mask = mask)
plt.axis("off")

plt.imshow(np.hstack([median, result]))
```

```python
brancozo = cv2.imread('LOTE_10.png')

plt.axis("off")
brancozo = cv2.cvtColor(brancozo, cv2.COLOR_BGR2RGB)
plt.imshow(brancozo)
plt.show()
```

```python
hsv = cv2.cvtColor(brancozo, cv2.COLOR_BGR2HSV)

lower = np.array([94, 20, 2])
upper = np.array([126, 255, 255])

mask1 = cv2.inRange(hsv, lower, upper)
plt.imshow(mask1)
plt.show()
```

```python
result = cv2.bitwise_and(brancozo ,brancozo, mask = mask1)
plt.imshow(np.hstack([brancozo, result]))
```

```python
plt.axis('off')
plt.imshow(result)
```

```python
seed_pixel = np.count_nonzero(result)
seed_pixel
```

```python
result_seed = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(result_seed, cv2.COLOR_BGR2HSV)

lower = np.array([0,0,168])
upper = np.array([172,111,255])

mask2 = cv2.inRange(hsv, lower, upper)
plt.axis("off")
plt.imshow(mask2)
plt.show()
```

```python
result_seed = cv2.cvtColor(result_seed, cv2.COLOR_BGR2RGB)

result_seed = cv2.bitwise_and(result_seed ,result_seed, mask = mask2)
plt.axis("off")

plt.imshow(result_seed)
```

```python
seed_pixel_brancozo = np.count_nonzero(result_seed)
seed_pixel_brancozo
```

```python
plt.axis('off')
plt.imshow(np.hstack([result, result_seed]))
```

```python
percent = ((100*seed_pixel_brancozo)/seed_pixel)
print('Porcentagem de Branco Leitoso: ', percent, '%')
```
