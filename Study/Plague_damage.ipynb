```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
target = './testetetrazolio.jpg'
input_image = cv.imread(target)
plt.axis("off")
plt.imshow(input_image)
plt.show()
```

```python
grayImg = cv.cvtColor(input_image, cv.COLOR_BGR2GRAY)
blurImg = cv.GaussianBlur(grayImg,(9,9),0)
plt.axis("off")
plt.imshow(blurImg)
plt.show()
```

```python
blurImg = cv.threshold(blurImg,185,255,cv.THRESH_BINARY)[1]
plt.axis("off")
plt.imshow(blurImg)
plt.show()
```

```python
canny = cv.Canny(blurImg,100,200)
plt.axis("off")
plt.imshow(canny, cmap = "gray")
plt.show()
```

```python
blurImg_inv = cv.cvtColor(input_image, cv.COLOR_BGR2RGB)

blurImg_inv = cv.bitwise_and(blurImg_inv ,blurImg_inv, mask = blurImg)
plt.axis("off")

plt.imshow(blurImg_inv)
```

```python
plt.imshow(np.hstack([input_image, blurImg_inv]))
```
