```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
feijao = cv2.imread('feijao.png')
plt.imshow(feijao)
plt.show()
```

# Foto 1

```python
feijao = cv2.cvtColor(feijao, cv2.COLOR_BGR2RGB)
plt.imshow(feijao)
plt.show()
```

# EXEMPLO 1 - Deteco Vermelho

```python
feijao = cv2.cvtColor(feijao, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(feijao, cv2.COLOR_BGR2HSV)

lower = np.array([0, 50, 50])
upper = np.array([10, 255, 255])

mask = cv2.inRange(hsv, lower, upper)
plt.imshow(mask)
plt.show()
```

```python
feijao = cv2.cvtColor(feijao, cv2.COLOR_BGR2RGB)

result = cv2.bitwise_and(feijao ,feijao, mask = mask)
plt.imshow(np.hstack([feijao, result]))
```

```python
feijao = cv2.cvtColor(feijao, cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(feijao, cv2.COLOR_BGR2HSV)


lower = np.array([0,0,168])
upper = np.array([172,111,255])

mask = cv2.inRange(hsv, lower, upper)
plt.imshow(mask)
plt.show()
```

# Exemplo 3 - Deteco de Vermelho

```python
on_fej = cv2.cvtColor(on_fej , cv2.COLOR_BGR2RGB)

hsv = cv2.cvtColor(on_fej , cv2.COLOR_BGR2HSV)

lower = np.array([0,100, 100])
upper = np.array([10, 255, 255])

mask = cv2.inRange(hsv, lower, upper)
plt.imshow(mask)
plt.show()
```

```python
on_fej = cv2.cvtColor(on_fej, cv2.COLOR_BGR2RGB)

result = cv2.bitwise_and(on_fej ,on_fej, mask = mask)
plt.imshow(result)
```

```python
totais_pixel = result.reshape((np.shape(result)[0]*np.shape(result)[1],3))
totais_pixel.shape
```

```python
red_pixel = np.count_nonzero(result)
red_pixel
```

# Porcentagem de Vermelho

```python
percent = ((100*red_pixel)/totais_pixel.shape[0])
print('Porcentagem de Vermelho: ', percent, '%')
```
